## 1. Datenbereinigung Unternehmen 1–50

```python
import os
os.getcwd()
```

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Ergebnisse_1_50.xlsx")
```

```python
id_col = "Unternehmensname Latin alphabet"
```

```python
# Spalten mit USD droppen
```

```python
usd_cols = df.columns[df.columns.str.contains("USD", case=False, na=False)]
df = df.drop(columns=usd_cols)
```

```python
# Spalten mit tsd droppen
```

```python
tsd_cols = df.columns[df.columns.str.contains("tsd", case=False, na=False)]
df = df.drop(columns=tsd_cols)
```

```python
# Duplikate droppen
```

```python
duplikate = df.columns[df.columns.str.contains(r"\.\d+$")]
df = df.drop(columns=duplikate)
```

```python
# Spalten mit Letztes verf. Jahr droppen
```

```python
df = df.loc[:, ~df.columns.str.contains("Letztes verf. Jahr", case=False)]
```

```python
# n.v. mit NA ersetzen
```

```python
df = df.replace("n.v.", pd.NA)
```

```python
bilanzsumme_cols = [c for c in df.columns if "Bilanzsumme Aktiva" in c]
umlaufvermögen_cols = [c for c in df.columns if "Umlaufvermögen" in c]
zahlungsmittel_cols = [c for c in df.columns if "Zahlungsmittel & Zahlungsmitteläquivalente" in c]
kurz_verb_cols = [c for c in df.columns if "Kurzfristige Verbindlichkeiten" in c]
lang_verb_cols = [c for c in df.columns if "Langfristige Verbindlichkeiten" in c]
EK_cols = [c for c in df.columns if "Eigenkapital\nmil CHF" in c]
umsatz_cols = [c for c in df.columns if "Umsatz" in c]
EBIT_cols = [c for c in df.columns if "Betriebsgewinn/-verlust" in c]
finanzaufwendungen_cols = [c for c in df.columns if "Finanzaufwendungen" in c]
gewinn_verlust_cols = [c for c in df.columns if "Gewinn/Verlust vor Steuern" in c]
ROE_cols = [c for c in df.columns if "ROE basierend auf Gewinn/Verlust vor Steuern" in c]
ROA_cols = [c for c in df.columns if "ROA basierend auf Gewinn/Verlust vor Steuern" in c]
FK_zu_EK_cols = [c for c in df.columns if "Fremd- zu Eigenkapital" in c]
EK_quote_cols = [c for c in df.columns if "Eigenkapitalquote" in c]
```

```python
# Wichtige Spalten erstellen
```

```python
# Bilanzsumme Aktiva
```

```python
bilanzsumme_long = df[[id_col] + bilanzsumme_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Bilanzsumme"
)

# Jahr aus Spaltennamen ziehen
bilanzsumme_long["Jahr"] = (
    bilanzsumme_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)
# Aufräumen
bilanzsumme_long = bilanzsumme_long.drop(columns="Spalte")
```

```python
# Eigenkapital
```

```python
EK_long = df[[id_col] + EK_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Eigenkapital"
)

# Jahr aus Spaltennamen ziehen
EK_long["Jahr"] = (
    EK_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)
# Aufräumen
EK_long = EK_long.drop(columns="Spalte")
```

```python
# ROA
```

```python
ROA_long = df[[id_col] + ROA_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="ROA"
)

# Jahr aus Spaltennamen ziehen
ROA_long["Jahr"] = (
    ROA_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)
# Aufräumen
ROA_long = ROA_long.drop(columns="Spalte")
```

```python
# Kurzfristige Verbindlichkeiten 
```

```python
kurz_verb_long = df[[id_col] + kurz_verb_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Kurzfristige_Verbindlichkeiten"
)

kurz_verb_long["Jahr"] = (
    kurz_verb_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)

kurz_verb_long = kurz_verb_long.drop(columns="Spalte")
```

```python
# Langfristige Verbindlichkeiten 
```

```python
lang_verb_long = df[[id_col] + lang_verb_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Langfristige_Verbindlichkeiten"
)

lang_verb_long["Jahr"] = (
    lang_verb_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)

lang_verb_long = lang_verb_long.drop(columns="Spalte")
```

```python
# Eigenkapital Quote
```

```python
EK_quote_long = df[[id_col] + EK_quote_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="EK_Quote"
)

EK_quote_long["Jahr"] = (
    EK_quote_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)

EK_quote_long = EK_quote_long.drop(columns="Spalte")
```

```python
# Umsatz
```

```python
umsatz_long = df[[id_col] + umsatz_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Umsatz"
)

umsatz_long["Jahr"] = (
    umsatz_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)

umsatz_long = umsatz_long.drop(columns="Spalte")
```

```python
# Zahlungsmittel
```

```python
zahlungsmittel_long = df[[id_col] + zahlungsmittel_cols].melt(
    id_vars=id_col,
    var_name="Spalte",
    value_name="Zahlungsmittel"
)

# Jahr aus Spaltennamen ziehen
zahlungsmittel_long["Jahr"] = (
    zahlungsmittel_long["Spalte"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)
# Aufräumen
zahlungsmittel_long = zahlungsmittel_long.drop(columns="Spalte")
```

```python
panel = (
    bilanzsumme_long
    .merge(zahlungsmittel_long, on=[id_col, "Jahr"], how="outer")
    .merge(ROA_long,            on=[id_col, "Jahr"], how="outer")
    .merge(EK_long,             on=[id_col, "Jahr"], how="outer")
    .merge(kurz_verb_long,      on=[id_col, "Jahr"], how="outer")
    .merge(lang_verb_long,      on=[id_col, "Jahr"], how="outer")
    .merge(EK_quote_long,       on=[id_col, "Jahr"], how="outer")
    .merge(umsatz_long,         on=[id_col, "Jahr"], how="outer")
)

Unternehmensdaten_1_50 = panel.sort_values([id_col, "Jahr"]).reset_index(drop=True)
```

```python
# Unternehmens- und Zinsdaten zusammenfügen
df_merged = Unternehmen.merge(
    Zins,
    on="Jahr",
    how="left"
)
```

```python
df_merged["Bilanzsumme"] = pd.to_numeric(
    df_merged["Bilanzsumme"],
    errors="coerce"
)
```

```python
# Zinswernde-Dummy erstellen
df_merged["Zinswende_Dummy"] = (df_merged["Jahr"] >= 2022).astype(int)
```

```python
# Variablen Berechnen
```

```python
df_merged["Cash_Ratio"] = (
    df_merged["Zahlungsmittel"] / df_merged["Bilanzsumme"]
)
```

```python
df_merged["Groesse_Log_Bilanzsumme"] = np.log(df_merged["Bilanzsumme"])
```

```python
company_col = "Unternehmensname Latin alphabet"
pre_years = range(2016, 2022)  # 2016–2021
```

```python
df_merged["Leverage"] = (
    (df_merged["Kurzfristige_Verbindlichkeiten"]
     + df_merged["Langfristige_Verbindlichkeiten"])
    / df_merged["Bilanzsumme"]
)
```

```python
leverage_exante = (
    df_merged[df_merged["Jahr"].isin(pre_years)]
    .groupby(company_col)["Leverage"]
    .mean()
)
```

```python
df_merged["Leverage_exante"] = df_merged[company_col].map(leverage_exante)
```

```python
liquidity_exante = (
    df_merged[df_merged["Jahr"].isin(pre_years)]
    .groupby(company_col)["Cash_Ratio"]
    .mean()
)
```

```python
df_merged["Liquidity_exante"] = df_merged[company_col].map(liquidity_exante)
```

```python
df_merged["STDebtShare"] = (
    df_merged["Kurzfristige_Verbindlichkeiten"] /
    (df_merged["Kurzfristige_Verbindlichkeiten"]
     + df_merged["Langfristige_Verbindlichkeiten"])
)
```

```python
stdebt_exante = (
    df_merged[df_merged["Jahr"].isin(pre_years)]
    .groupby(company_col)["STDebtShare"]
    .mean()
)
```

```python
df_merged["STDebtShare_exante"] = df_merged[company_col].map(stdebt_exante)
```
